```python
# Primeiro, vamos garantir que temos as ferramentas certas
# O comando "!" permite executar comandos do terminal dentro do notebook
# pandas: a biblioteca mais popular para manipula√ß√£o de dados em Python
# duckdb-engine: o "motor" que nos permite usar SQL em cima dos nossos dados
#   caso voc√™ se interesse em saber mais o que √© o duckdb acesse: https://duckdb.org/

!pip install pandas duckdb-engine
```

```python
# Agora que as bibliotecas est√£o instaladas, vamos import√°-las para o nosso c√≥digo.
# Assim, podemos usar as fun√ß√µes que elas oferecem.

import pandas as pd
import duckdb

print("Bibliotecas importadas com sucesso!")
```

```python
# Aqui est√° a vari√°vel com a fonte de dados que vamos usar. √â um link direto para um arquivo CSV.
url_dados = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'

# Vamos usar o pandas para ler o arquivo CSV a partir do link e carreg√°-lo em uma estrutura chamada DataFrame.
# Pense em um DataFrame como uma tabela ou uma planilha do Excel dentro do Python.
df_pinguins = pd.read_csv(url_dados)
```

```python
# Vamos criar uma conex√£o com o DuckDB.
# O par√¢metro ':memory:' significa que o banco de dados vai rodar inteiramente na mem√≥ria RAM
con = duckdb.connect(database=':memory:')

# Agora, o passo mais importante: vamos registrar nosso DataFrame 'df_pinguins' como uma tabela SQL.
# A partir de agora, o DuckDB vai "enxergar" nosso DataFrame como uma tabela chamada 'pinguins'.
# √â nesta tabela 'pinguins' que vamos fazer todas as nossas consultas SQL.
con.register('pinguins', df_pinguins)

print("Ambiente SQL pronto! A tabela 'pinguins' foi criada na mem√≥ria.")
```

```python
# Para facilitar nossa vida, vamos criar uma fun√ß√£o que executa uma query SQL
# e j√° retorna o resultado formatado como uma bela tabela do pandas.
# Assim, n√£o precisamos repetir o mesmo c√≥digo v√°rias vezes.

def executar_query(query_sql):
  """
  Recebe uma string com um comando SQL, executa no DuckDB e retorna um DataFrame do pandas.
  """
  return con.execute(query_sql).fetchdf()

print("Fun√ß√£o auxiliar 'executar_query' criada com sucesso.")
```

```python
# √â sempre uma boa pr√°tica olhar as primeiras linhas dos seus dados para entender como eles s√£o.

query_all = """
  SELECT * FROM pinguins LIMIT 10;
"""

print("Visualiza√ß√£o geral:")
reesultado_all = executar_query(query_all)
display(reesultado_all)
```

```python
# Exerc√≠cio 1: Mostre apenas as colunas de esp√©cie, ilha e sexo dos pinguins, limitando a 10 resultados.

query_ex1 = """
  SELECT SPECIES, ISLAND, SEX FROM pinguins LIMIT 10;
"""

print("Resultado do Exerc√≠cio 1:")
resultado_ex1 = executar_query(query_ex1)
display(resultado_ex1)
```

```python
# Exerc√≠cio 2: Mostre todos os dados dos pinguins da esp√©cie "Adelie" que vivem na ilha "Torgersen".

query_ex2 = """
  SELECT *
  FROM pinguins
  WHERE species='Adelie' AND ISLAND='Torgersen';

"""

print("Resultado do Exerc√≠cio 2:")
resultado_ex2 = executar_query(query_ex2)
display(resultado_ex2)
```

```python
# Exerc√≠cio 3: Conte quantos pinguins existem em cada uma das ilhas.

query_ex3 = """
  Select island, count(island) as Pinguins_por_ilha
  from pinguins
  group by island

"""

print("Resultado do Exerc√≠cio 3:")
resultado_ex3 = executar_query(query_ex3)
display(resultado_ex3)
```

```python
# Exerc√≠cio 4: Calcule o peso m√©dio (body_mass_g) dos pinguins para cada esp√©cie.
# Ordene o resultado do maior peso m√©dio para o menor.

query_ex4 = """
Select species, AVG(body_mass_g) as peso_medio
from pinguins
group by species
order by species

"""

print("Resultado do Exerc√≠cio 4:")
resultado_ex4 = executar_query(query_ex4)
display(resultado_ex4)
```

```python
# Exerc√≠cio 5 (Desafio): Encontre a esp√©cie e a ilha do pinguim com o maior comprimento de bico (bill_length_mm).
# Dica: Voc√™ pode precisar usar ORDER BY e LIMIT juntos para encontrar o "maior".

query_ex5 = """
Select species, island, Max(bill_length_mm) as Maior_comprimento_bico
from pinguins
group by species, island
order by Maior_comprimento_bico desc
LIMIT 1
"""

print("Resultado do Exerc√≠cio 5 (Desafio):")
resultado_ex5 = executar_query(query_ex5)
display(resultado_ex5)
```

```python
# Exerc√≠cio 6: Encontre todos os pinguins machos ('male') que t√™m o comprimento da nadadeira (flipper_length_mm) maior ou igual a 220 mm.
# Mostre a esp√©cie, a ilha e o comprimento da nadadeira.

query_ex6 = """
Select species, island, flipper_length_mm
from pinguins
where sex='male' and flipper_length_mm>219
"""

print("Resultado do Exerc√≠cio 6:")
resultado_ex6 = executar_query(query_ex6)
display(resultado_ex6)
```

```python
# Exerc√≠cio 7: Para cada sexo, qual √© a m√©dia do comprimento do bico (bill_length_mm) e a m√©dia da profundidade do bico (bill_depth_mm)?
# Ignore os registros onde o sexo n√£o est√° preenchido (√© NULO).

query_ex7 = """
select sex, avg(bill_length_mm),avg(bill_depth_mm)
from pinguins
where sex is not null
group by sex;
"""

print("Resultado do Exerc√≠cio 7:")
resultado_ex7 = executar_query(query_ex7)
display(resultado_ex7)
```

```python
# Exerc√≠cio 8: Quantos pinguins da esp√©cie "Gentoo" foram registrados em cada ano?
# Ordene os resultados pelo ano.

query_ex8 = """
select year, count(species) as total
from pinguins
where species='Gentoo'
group by year


"""

print("Resultado do Exerc√≠cio 8:")
resultado_ex8 = executar_query(query_ex8)
display(resultado_ex8)
```

```python
# Exerc√≠cio 9: Crie uma nova coluna chamada "classificacao_peso" para os pinguins da esp√©cie "Chinstrap".
# Se o peso (body_mass_g) for maior que 4000, classifique como 'Pesado'.
# Se for menor ou igual a 4000, classifique como 'Leve'.
# Mostre a ilha, o peso e a nova coluna de classifica√ß√£o.

query_ex9 = """
select island, body_mass_g as peso, if(body_mass_g>4000,'PESADO', 'LEVE') as classifica√ß√£o
from pinguins
where species='Chinstrap'

"""

print("Resultado do Exerc√≠cio 9:")
resultado_ex9 = executar_query(query_ex9)
display(resultado_ex9)
```

```python
# Exerc√≠cio 10 (Desafio Intermedi√°rio): Encontre o peso (body_mass_g) do pinguim mais pesado e do mais leve para cada combina√ß√£o de ilha e esp√©cie.
# Sua tabela final deve ter as colunas: ilha, especie, peso_maximo e peso_minimo.

query_ex10 = """
select island, species, max(body_mass_g) as peso_maximo, min(body_mass_g) as peso_minimo
from pinguins
group by island, species
"""

print("Resultado do Exerc√≠cio 10:")
resultado_ex10 = executar_query(query_ex10)
display(resultado_ex10)
```
